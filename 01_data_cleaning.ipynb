```python
# converted CustomerID to integer.
df['CustomerID'] = df['CustomerID'].astype(int)
df.info()
```

```python
# 🔥 STEP 12 – Feature Engineering (Where Business Starts)
```

```python
# Revenue = Quantity × UnitPrice
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df['Revenue'].describe()
```

```python
# 🔥 STEP 13 – Order-Level Aggregation
```

```python
order_df = df.groupby(['InvoiceNo', 'CustomerID', 'InvoiceDate', 'Country'])['Revenue'].sum().reset_index()

order_df.head()
```

```python
order_df.shape
```

```python
# 🔥 STEP 14 – Basic Order-Level Insights
```

```python
order_df['Revenue'].describe()
```

```python
order_df['CustomerID'].nunique()
```

```python
# 🔥 STEP 15 – Customer-Level Aggregation
```

```python
customer_df = order_df.groupby('CustomerID').agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Orders=('InvoiceNo', 'count'),
    First_Purchase=('InvoiceDate', 'min'),
    Last_Purchase=('InvoiceDate', 'max')
).reset_index()

customer_df.head()
```

```python
customer_df.shape
```

```python
# 🔥 STEP 16 – Create Recency Metric
```

```python
max_date = order_df['InvoiceDate'].max()
max_date
```

```python
customer_df['Recency_Days'] = (max_date - customer_df['Last_Purchase']).dt.days
customer_df.head()
```

```python
# 🔥 STEP 17 – Confirm Frequency & Monetary
```

```python
customer_df[['Total_Orders', 'Total_Revenue', 'Recency_Days']].describe()
```

```python
# # 🔥 STEP 18 – Create RFM Scores
# We will create:
# R_Score (lower Recency = better)
# F_Score (higher Frequency = better)
# M_Score (higher Monetary = better)
```

```python
# Recency Score (reverse order)
customer_df['R_Score'] = pd.qcut(customer_df['Recency_Days'], 5, labels=[5,4,3,2,1])

# Frequency Score
customer_df['F_Score'] = pd.qcut(customer_df['Total_Orders'].rank(method='first'), 5, labels=[1,2,3,4,5])

# Monetary Score
customer_df['M_Score'] = pd.qcut(customer_df['Total_Revenue'], 5, labels=[1,2,3,4,5])
```

```python
customer_df[['CustomerID','R_Score','F_Score','M_Score']].head()
```

```python
# 🔥 STEP 19 – Create Combined RFM Score
```

```python
customer_df['RFM_Score'] = (
    customer_df['R_Score'].astype(str) +
    customer_df['F_Score'].astype(str) +
    customer_df['M_Score'].astype(str)
)

customer_df[['CustomerID','RFM_Score']].head()
```

```python
customer_df['RFM_Score'].value_counts().head(10)
```

```python
# 🔥 STEP 20 – Create RFM Segments
```

```python
def segment_customer(row):
    if row['RFM_Score'] == '555':
        return 'Champions'
    elif row['R_Score'] >= 4 and row['F_Score'] >= 4:
        return 'Loyal Customers'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 4:
        return 'At Risk'
    elif row['R_Score'] >= 4 and row['F_Score'] <= 2:
        return 'Potential Loyalist'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2:
        return 'Lost'
    else:
        return 'Others'

customer_df['Segment'] = customer_df.apply(segment_customer, axis=1)

customer_df['Segment'].value_counts()
```

**🎯 Now We Validate Pareto Principle**

We test:
Do top customers generate most revenue?

```python
# Sort customers by revenue
top_customers = customer_df.sort_values(by='Total_Revenue', ascending=False)

# Calculate cumulative revenue percentage
top_customers['Revenue_%'] = top_customers['Total_Revenue'] / top_customers['Total_Revenue'].sum()
top_customers['Cumulative_%'] = top_customers['Revenue_%'].cumsum()

top_customers.head()
```

```python
top_customers[top_customers['Cumulative_%'] <= 0.8].shape
```

🚀 STEP 2 — Create Date Dimension in Python

```python
date_df = pd.DataFrame({
    'FullDate': pd.to_datetime(order_df['InvoiceDate'].dt.date.unique())
})

date_df['DateKey'] = date_df['FullDate'].dt.strftime('%Y%m%d').astype(int)
date_df['Year'] = date_df['FullDate'].dt.year
date_df['Month'] = date_df['FullDate'].dt.month
date_df['MonthName'] = date_df['FullDate'].dt.strftime('%B')
date_df['Quarter'] = date_df['FullDate'].dt.quarter

date_df.head()
```
